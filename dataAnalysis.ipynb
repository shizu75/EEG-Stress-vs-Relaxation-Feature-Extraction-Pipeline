```python
import mne_features.univariate as mne_f
import numpy as np
import pandas as pd


def time_series_features(data):
    '''
    Computes the features variance, RMS, and peak-to-peak amplitude for each 1-second window.

    Args:
        data (ndarray): EEG data (n_patients x 3200 samples).

    Returns:
        ndarray: Computed features for each window.
    '''

    n_patients, n_samples = data.shape
    window_size = 128  # 1 second, since the sampling frequency is 128 Hz
    num_windows = n_samples // window_size  # 25 windows for 25 seconds

    features_per_patient = 3  # variance, RMS, peak-to-peak amplitude
    all_features = []

    for i, patient_data in enumerate(data):
        patient_features = []
        for j in range(num_windows):
            start = j * window_size
            end = (j + 1) * window_size
            window_data = patient_data[start:end]

            variance = mne_f.compute_variance(window_data)
            rms = mne_f.compute_rms(window_data)
            ptp_amp = mne_f.compute_ptp_amp(window_data)
            features = np.concatenate([variance, rms, ptp_amp])
            patient_features.append(features)

        all_features.append(np.array(patient_features).flatten())

    return np.array(all_features)


def freq_band_features(data, freq_bands):
    '''
    Computes the frequency bands delta, theta, alpha, beta, and gamma for each 1-second window
    and includes max frequency and power spectrum features.

    Args:
        data (ndarray): EEG data (n_patients x 3200 samples).
        freq_bands (ndarray): The frequency bands to compute.

    Returns:
        ndarray: Computed features for each window.
    '''

    n_patients, n_samples = data.shape
    window_size = 128  # 1 second, since the sampling frequency is 128 Hz
    num_windows = n_samples // window_size  # 25 windows for 25 seconds

    features_per_patient = len(freq_bands) - 1  # Number of frequency bands
    all_features = []

    for i, patient_data in enumerate(data):
        patient_features = []
        for j in range(num_windows):
            start = j * window_size
            end = (j + 1) * window_size
            window_data = patient_data[start:end]

            # Power spectrum for the window
            psd = mne_f.compute_pow_freq_bands(128, window_data, freq_bands=freq_bands)

            # Max frequency for each band across the whole EEG signal
            psd_all_bands = mne_f.compute_pow_freq_bands(128, patient_data, freq_bands=freq_bands)
            max_freq = []
            for band in range(len(freq_bands) - 1):
                band_start_idx = np.searchsorted(freq_bands, freq_bands[band])
                band_end_idx = np.searchsorted(freq_bands, freq_bands[band+1])
                band_data = psd_all_bands[band_start_idx:band_end_idx]
                max_freq.append(band_data.argmax())  # Find the frequency with max power

            patient_features.extend(psd)
            patient_features.extend(max_freq)  # Append max frequencies for each band

        all_features.append(np.array(patient_features).flatten())

    return np.array(all_features)


def hjorth_features(data):
    '''
    Computes the Hjorth mobility (spectral) and Hjorth complexity (spectral) for each 1-second window.

    Args:
        data (ndarray): EEG data (n_patients x 3200 samples).

    Returns:
        ndarray: Computed features for each window.
    '''

    n_patients, n_samples = data.shape
    window_size = 128  # 1 second, since the sampling frequency is 128 Hz
    num_windows = n_samples // window_size  # 25 windows for 25 seconds

    features_per_patient = 2  # Hjorth mobility and complexity
    all_features = []

    for i, patient_data in enumerate(data):
        patient_features = []
        for j in range(num_windows):
            start = j * window_size
            end = (j + 1) * window_size
            window_data = patient_data[start:end]

            mobility_spect = mne_f.compute_hjorth_mobility_spect(128, window_data)
            complexity_spect = mne_f.compute_hjorth_complexity_spect(128, window_data)
            features = np.concatenate([mobility_spect, complexity_spect])
            patient_features.append(features)

        all_features.append(np.array(patient_features).flatten())

    return np.array(all_features)


def extract_features(data, freq_bands):
    '''
    Extracts all features and saves them in a DataFrame for each patient.

    Args:
        data (ndarray): EEG data (n_patients x 3200 samples).
        freq_bands (ndarray): The frequency bands to compute.

    Returns:
        pd.DataFrame: DataFrame containing all extracted features.
    '''

    time_features = time_series_features(data)
    freq_features = freq_band_features(data, freq_bands)
    hjorth_features_data = hjorth_features(data)

    # Concatenate all features into one matrix
    all_features = np.hstack([time_features, freq_features, hjorth_features_data])

    # Create column labels based on the number of features
    time_feature_labels = [f"TimeFeature_{i+1}" for i in range(time_features.shape[1])]
    freq_feature_labels = [f"FreqFeature_{i+1}" for i in range(freq_features.shape[1])]
    hjorth_feature_labels = [f"HjorthFeature_{i+1}" for i in range(hjorth_features_data.shape[1])]

    # Add labels for new frequency features (max frequency for each band)
    max_freq_labels = [f"MaxFreq_{band}" for band in ['Delta', 'Theta', 'Alpha', 'Beta', 'Gamma']]

    column_labels = time_feature_labels + freq_feature_labels + hjorth_feature_labels + max_freq_labels

    # Create a DataFrame
    df = pd.DataFrame(all_features, columns=column_labels)

    return df


# Example usage:
# Load your data (replace 'data.csv' with your actual file path)
data = np.loadtxt('/content/Stressed.csv', delimiter=',')  # n_patients x 3200 samples
freq_bands = np.array([0.5, 4, 8, 12, 30, 45])  # Example frequency bands: Delta, Theta, Alpha, Beta, Gamma

# Extract features
feature_df = extract_features(data, freq_bands)

# Save the features to a CSV file
feature_df.to_csv('eeg_features.csv', index=False)
```

```python
import pandas as pd
import numpy as np
from scipy.signal import welch

# Load data
data = np.loadtxt('/content/Stressed.csv', delimiter=',')
pat, sam = data.shape

fs = 128  # Sampling frequency
window = int(sam/fs)  # Number of windows, 25 windows of 1 second each

data_rows = []

# Define frequency bands
freq_bands = {
    'Delta': (0.1, 3),
    'Theta': (4, 7),
    'Alpha': (8, 12),
    'Beta': (13, 31),
    'Gamma': (32, 44)
}

for i in range(pat):
    row_data = {}
    d = data[i, :]  # Select patient data

    for j in range(window):
        start = j * fs
        end = (j + 1) * fs
        window_data = d[start:end]

        # Compute power spectral density using Welch's method
        f, psd = welch(window_data, fs=fs, nperseg=fs)

        # Extract frequency with maximum power for each band
        delta_idx = np.logical_and(f >= freq_bands['Delta'][0], f <= freq_bands['Delta'][1])
        theta_idx = np.logical_and(f >= freq_bands['Theta'][0], f <= freq_bands['Theta'][1])
        alpha_idx = np.logical_and(f >= freq_bands['Alpha'][0], f <= freq_bands['Alpha'][1])
        beta_idx = np.logical_and(f >= freq_bands['Beta'][0], f <= freq_bands['Beta'][1])
        gamma_idx = np.logical_and(f >= freq_bands['Gamma'][0], f <= freq_bands['Gamma'][1])

        max_delta_freq = f[delta_idx][np.argmax(psd[delta_idx])]
        max_theta_freq = f[theta_idx][np.argmax(psd[theta_idx])]
        max_alpha_freq = f[alpha_idx][np.argmax(psd[alpha_idx])]
        max_beta_freq = f[beta_idx][np.argmax(psd[beta_idx])]
        max_gamma_freq = f[gamma_idx][np.argmax(psd[gamma_idx])]

        row_data[f'window_{j}_delta'] = np.sum(psd[delta_idx])
        row_data[f'window_{j}_theta'] = np.sum(psd[theta_idx])
        row_data[f'window_{j}_alpha'] = np.sum(psd[alpha_idx])
        row_data[f'window_{j}_beta'] = np.sum(psd[beta_idx])
        row_data[f'window_{j}_gamma'] = np.sum(psd[gamma_idx])

        row_data[f'window_{j}_max_delta_freq'] = max_delta_freq
        row_data[f'window_{j}_max_theta_freq'] = max_theta_freq
        row_data[f'window_{j}_max_alpha_freq'] = max_alpha_freq
        row_data[f'window_{j}_max_beta_freq'] = max_beta_freq
        row_data[f'window_{j}_max_gamma_freq'] = max_gamma_freq

    data_rows.append(row_data)

# Convert to DataFrame and save to CSV
df = pd.DataFrame(data_rows)
df.to_csv('eeg_features_Stressed_Frequency.csv', index=False)

print(df)
```

```python
import pandas as pd
import numpy as np
import mne_features.univariate as mne_f

data = np.loadtxt('/content/Relaxed.csv', delimiter=',')
pat, sam = data.shape

fs = 128
window = int(sam/fs)

data_rows = []

for i in range(pat):
    row_data = {}
    d = data[i,:]
    for j in range(window):
        start = j * fs
        end = (j + 1) * fs
        window_data = d[start:end]

        # Reshape window_data to (1, -1) to represent a single channel
        window_data = window_data.reshape(1, -1)  # Reshaping to (1, 128) for single-channel data

        app_entropy = mne_f.compute_app_entropy(window_data)
        samp_entropy = mne_f.compute_samp_entropy(window_data)
        spect_entropy = mne_f.compute_spect_entropy(fs, window_data)
        svd_entropy = mne_f.compute_svd_entropy(window_data)

        # Store the features with unique column names
        row_data[f'window_{j}_app_entropy'] = app_entropy
        row_data[f'window_{j}_samp_entropy'] = samp_entropy
        row_data[f'window_{j}_spect_entropy'] = spect_entropy
        row_data[f'window_{j}_svd_entropy'] = svd_entropy

    data_rows.append(row_data)

# Create DataFrame
df = pd.DataFrame(data_rows)

# Save DataFrame to CSV
df.to_csv('/content/relaxEntropyfeatures.csv', index=False)

print("Features saved to 'csv'.")
```

```python
import pandas as pd
import numpy as np
from scipy.signal import butter, filtfilt, hilbert

# Define frequency bands
freq_bands = {
    'Delta': (0.5, 4),
    'Theta': (4, 8),
    'Alpha': (8, 12),
    'Beta': (12, 30),
    'Gamma': (30, 45)
}

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data)

# Load data
data = np.loadtxt('/content/Stressed.csv', delimiter=',')
pat, sam = data.shape

fs = 128  # Sampling frequency
window = int(sam / fs)

data_rows = []

for i in range(pat):
    row_data = {}
    d = data[i, :]
    for j in range(window):
        start = j * fs
        end = (j + 1) * fs
        window_data = d[start:end]

        # Process each frequency band
        for band, (low, high) in freq_bands.items():
            # Apply bandpass filter
            filtered_data = bandpass_filter(window_data, low, high, fs)

            # Compute the envelope using the Hilbert transform
            envelope = np.abs(hilbert(filtered_data))

            # Find max power and corresponding index
            max_power = np.max(envelope)
            max_index = np.argmax(envelope)
            max_freq = low + (high - low) * (max_index / len(envelope))

            # Store results in row_data
            row_data[f'window_{j}_{band}_max_power'] = max_power
            row_data[f'window_{j}_{band}_max_freq'] = max_freq

    data_rows.append(row_data)

# Create DataFrame
df = pd.DataFrame(data_rows)

# Save DataFrame to CSV
print(df)

print("Features saved to 'relaxed_band_features_hilbert.csv'.")
```

```python
from scipy.signal import welch
import pandas as pd
import numpy as np

# Define frequency bands
freq_bands = {
    'Delta': (0.5, 4),
    'Theta': (4, 8),
    'Alpha': (8, 12),
    'Beta': (12, 30),
    'Gamma': (30, 45)
}

# Load data
data = np.loadtxt('/content/Relaxed.csv', delimiter=',')
pat, sam = data.shape
fs = 128  # Sampling frequency
window_size = fs * 2  # 2-second windows
overlap = window_size // 2

data_rows = []

for i in range(pat):
    row_data = {}
    d = data[i, :]

    for j in range(0, sam - window_size + 1, window_size - overlap):
        window_data = d[j:j + window_size]

        # Compute Welch PSD
        freqs, psd = welch(window_data, fs, nperseg=window_size)

        # Compute band powers
        for band, (low, high) in freq_bands.items():
            idx_band = np.logical_and(freqs >= low, freqs < high)
            band_power = np.sum(psd[idx_band])
            max_power = np.max(psd[idx_band])  # Max power within band
            max_freq = freqs[idx_band][np.argmax(psd[idx_band])]  # Frequency of max power

            row_data[f'window_{j}_{band}_power'] = band_power
            row_data[f'window_{j}_{band}_max_power'] = max_power
            row_data[f'window_{j}_{band}_max_freq'] = max_freq

    data_rows.append(row_data)

# Create DataFrame
df = pd.DataFrame(data_rows)

# Save to CSV
df.to_csv('/content/freqstress.csv', index=False)
print("Features saved to 'enhanced_features.csv'.")
```

```python
import pywt
import numpy as np
import pandas as pd
from scipy.signal import welch

# Define frequency bands
freq_bands = {
    'Delta': (0.5, 4),
    'Theta': (4, 8),
    'Alpha': (8, 12),
    'Beta': (12, 30),
    'Gamma': (30, 45)
}

# Wavelet thresholding function
def wavelet_denoise(signal, wavelet='bior2.2', level=4, threshold_method='soft'):
    coeffs = pywt.wavedec(signal, wavelet, level=level)
    sigma = np.median(np.abs(coeffs[-1])) / 0.6745
    threshold = sigma * np.sqrt(2 * np.log(len(signal)))

    denoised_coeffs = [pywt.threshold(c, threshold, mode=threshold_method) for c in coeffs]
    denoised_signal = pywt.waverec(denoised_coeffs, wavelet)

    return denoised_signal

# Load EEG data
data = np.loadtxt('/content/Relaxed.csv', delimiter=',')
pat, sam = data.shape
fs = 128  # Sampling frequency
window_size = fs * 2  # 2-second windows
overlap = window_size // 2

data_rows = []

for i in range(pat):
    row_data = {}
    d = data[i, :]

    for j in range(0, sam - window_size + 1, window_size - overlap):
        window_data = d[j:j + window_size]

        # Apply wavelet denoising
        denoised_data = wavelet_denoise(window_data, wavelet='bior2.2', level=4)

        # Compute Welch PSD
        freqs, psd = welch(denoised_data, fs, nperseg=window_size)

        # Compute band powers and max values
        for band, (low, high) in freq_bands.items():
            idx_band = np.logical_and(freqs >= low, freqs < high)
            band_power = np.sum(psd[idx_band])
            max_power = np.max(psd[idx_band])  # Max power within band
            max_freq = freqs[idx_band][np.argmax(psd[idx_band])]  # Frequency of max power

            row_data[f'window_{j}_{band}_power'] = band_power
            row_data[f'window_{j}_{band}_max_power'] = max_power
            row_data[f'window_{j}_{band}_max_freq'] = max_freq

    data_rows.append(row_data)

# Create DataFrame
df = pd.DataFrame(data_rows)

# Save to CSV
print(df)
print("Features saved to 'wavelet_features.csv'.")
```
